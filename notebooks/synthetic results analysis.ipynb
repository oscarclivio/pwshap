```python
import pandas as pd
import numpy as np
```

```python
for filename in ['bias_no_confounders','bias_one_confounder','bias_two_confounders','mediation_no_mediators','mediation_one_mediator','mediation_two_mediators']:
    df = pd.read_csv(f'../outputs/{filename}.csv')
    df = df[[col for col in df.columns if 'Unnamed' not in col]]
    print(filename)
    print('Means :')
    print(df.mean().round(3))
    print('Stds :')
    print((df.std() / np.sqrt(25)).round(3))
    print('===================================')
```
